```python
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np

f = open('./logs/train7.log')

lines = [line.rstrip("\n") for line in f.readlines()]

iters = []
loss = []
avg_loss = []

fig, ax = plt.subplots()

for line in lines:
    if ':' in line and 'images' in line:
        loss.append(float(line.split(' ')[1].split(',')[0]))
        avg_loss.append(float(line.split(', ')[1].split(' ')[0]))
        iters.append(int(line.split(':')[0]))        

ax.plot(iters, loss)
plt.xlabel('iters')
plt.ylabel('loss')
plt.grid()

ticks = range(0, 250, 5)

# ax.set_yticks(ticks)
plt.show()

np.max(iters),loss[-1],np.min(loss)
```

```python
fig, ax = plt.subplots()

ax.plot(iters, avg_loss)
plt.xlabel('iters')
plt.ylabel('avg loss')
plt.grid()

ticks = range(0, 250, 5)

# ax.set_yticks(ticks)
plt.show()
np.max(iters),avg_loss[-1],np.min(avg_loss)
```
